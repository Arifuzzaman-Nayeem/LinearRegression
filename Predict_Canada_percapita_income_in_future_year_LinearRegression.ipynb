```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Canada_per_capitaincome.csv")

plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker='+' )
plt.show()
```

```python
#fitting linearregression model
lin_reg = LinearRegression()
inputs = df[['year']]
target = df[['per capita income (US$)']]
lin_reg.fit(inputs,target)
fit_result = (lin_reg.score(inputs,target),lin_reg.coef_,lin_reg.intercept_)
print ('score: ', fit_result[0], ' coeff: ', fit_result[1], ' intercept: ', fit_result[2])
#best fit draw 
plt.scatter(inputs,target,color='red',marker='+' )
plt.plot(inputs,lin_reg.predict(inputs),color= 'green')
plt.show()
#predict per_capita_income_in_2020
predict_value = lin_reg.predict([[2020]])
#insert_predicted_value_in_existng_Df
df.loc['47'] = [2020,predict_value[0][0]]
#Saving Df in CSV format
df.to_csv('Result-Canada_per_capitaincome_prediction(LinearRegression).csv', index=False)
```
